```python
# ソートのメソッド群

def bubble_sort(arr):
    change = True
    while change:
        change = False
        for i in range(len(arr) - 1):
            if arr[i][1] > arr[i + 1][1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
                change = True
    return arr

def select_pivot(buff, low, high):
    a = buff[low]
    b = buff[(low + high) / 2]
    c = buff[high]
    if a > b:
        tmp = a
        a = b
        b = tmp
    if b > c:
        b = c
        if a > b: b = a
    return b

def quick_sort(buff, low, high):
    pivot = buff[int((low + high) / 2)][1]
    i = low
    j = high
    while True:
        while pivot > buff[i][1]:
            i += 1
        while pivot < buff[j][1]:
            j -= 1
        if i >= j: 
            break
        temp = buff[i][1]
        buff[i][1] = buff[j][1]
        buff[j][1] = temp
        i += 1
        j -= 1
    if low < i - 1: 
        quick_sort(buff, low, i - 1)
    if high > j + 1: 
        quick_sort(buff, j + 1, high)
        
def _quick_sort(array, left, right):
    if left >= right:
        print("end")
        return

    # ピボットを決定
    pivot = median(array[left][1], array[(left + (right - left) // 2)][1], array[right][1])
    l = left
    r = right
    # ピボットを中心に分ける
    while True:
        # ピボットより小さい値を左、大きい値を右に移動する
        while array[l][1] < pivot:
            l += 1
            if l == r:
                break
        while pivot < array[r][1]:
            r -= 1
            if r ==0:
                break
        if (r >= l): 
            break

        # 見つかった要素を交換
        tmp = array[l][1]
        array[l][1] = array[r][1]
        array[r][1] = tmp
        l += 1
        r -= 1

    # 左右に分けた部分を再帰的にクイックソート
    _quick_sort(array, left, l - 1)
    _quick_sort(array, r + 1, right)

def quick_sort3(array):
    return _quick_sort(array, 0, len(array)-1)

def median(x, y, z):
    if x < y:
        return (y if y < z else z) if x < z else x
    else:
        return (x if x < z else z) if y < z else y

LIMIT = 10

def quick_sort2(buff, low, high):
    stack = []
    while True:
        if high - low <= LIMIT:
            if len(stack) == 0: break
            low, high = stack.pop()
        else:
            pivot = select_pivot(buff, low, high)
            i = low
            j = high
            while True:
                while pivot > buff[i]: i += 1
                while pivot < buff[j]: j -= 1
                if i >= j: break
                temp = buff[i]
                buff[i] = buff[j]
                buff[j] = temp
                i += 1
                j -= 1
            if i - low > high - j:
                stack.append((low, i - 1))
                low = j + 1
            else:
                stack.append((j + 1, high))
                high = i - 1
    insert_sort(buff)

def quick_sort4(target):
    if len(target) <= 1:
        return target

    pivot = target[100][1]
    left = []
    right = []
    for x in range(1, len(target)):
        if target[x][1] <= pivot:
            left.append(target[x][1])
        else:
            right.append(target[x][1])

    left = quick_sort4(left)
    right = quick_sort4(right)
    mid = [pivot]
    return left + mid + right

def shell_sort(buff):
    k = len(buff)
    gap = 1
    while gap < k / 9: gap = gap * 3 + 1
    while gap > 0:
        for i in range(gap, k):
            temp = buff[i]
            j = i - gap
            while j >= 0 and temp < buff[j]:
                buff[j + gap] = buff[j]
                j -= gap
            buff[j + gap] = temp
        gap /= 3
    return buff
```
