```python
def qsort(ls, left, right):

    i = left + 1
    k = right
    
    while(i < k):
        while((ls[left][1] > ls[i][1]) and (i < right)):
            i += 1
        while((ls[left][1] < ls[k][1]) and (k > left)):
            k -= 1
        if(i < k):
            ls[i], ls[k] = ls[k], ls[i]

    if(ls[left][1] > ls[k][1]):
        ls[left], ls[k] = ls[k], ls[left]

    if(left < k-1):
        qsort(ls, left, k-1)
    if(right > k+1):
        qsort(ls, k+1, right)
        
    return
```

```python
def quick_sort(buff, low, high):
    pivot = buff[int((low + high) / 2)][1]
    i = low
    j = high
    while True:
        while pivot < buff[i][1]:
            i += 1
        while pivot > buff[j][1]:
            j -= 1
        if i >= j: 
            break
        temp = buff[i][1]
        buff[i][1] = buff[j][1]
        buff[j][1] = temp
        i += 1
        j -= 1
    if low < i - 1: 
        quick_sort(buff, low, i - 1)
    if high > j + 1: 
        quick_sort(buff, j + 1, high)
```
